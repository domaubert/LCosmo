```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
import scipy.constants as pcst
plt.rcParams['figure.figsize'] = (10.0, 10.0) 
```

```python
data=np.loadtxt('bbn.txt',skiprows=1)
eta=data[:,0]
Yp=data[:,1]
D=data[:,2]
He3=data[:,3]
Li7=data[:,4]
```

```python
eta
```

```python
ee=np.logspace(-10.2,-8.9)
uu=np.logspace(-11,0)



plt.figure()


plt.subplot(311)
plt.plot(eta,Yp)
plt.fill_between(ee,ee*0+0.241,y2=ee*0+0.249,alpha=0.3,hatch='/')
plt.fill_betweenx(uu,uu*0+6.03e-10,x2=uu*0+6.15e-10,alpha=0.3)


plt.xscale('log')
plt.legend(['Yp'],fontsize=15)
plt.xlim([1e-10,1e-9])
plt.ylim([0.23,0.26])

plt.subplot(312)
plt.plot(eta,D)

plt.plot(eta,He3,linestyle='-.')
plt.fill_between(ee,ee*0+2.49e-5,y2=ee*0+2.57e-5,alpha=0.3,hatch='/')
plt.fill_betweenx(uu,uu*0+6.03e-10,x2=uu*0+6.15e-10,alpha=0.3)


plt.xlim([1e-10,1e-9])

plt.ylim([5e-6,5e-4])



plt.legend(['D/H','He3/H'],fontsize=15)



plt.xscale('log')
plt.yscale('log')

plt.subplot(313)
plt.plot(eta,Li7)
plt.fill_between(ee,ee*0+1.3e-10,y2=ee*0+1.9e-10,alpha=0.3,hatch='/')
plt.fill_betweenx(uu,uu*0+6.03e-10,x2=uu*0+6.15e-10,alpha=0.3)
plt.ylim([5e-11,1e-9])





plt.legend(['Li7/H'],fontsize=15)
plt.xlim([1e-10,1e-9])
plt.xlabel(r'rapport baryons/photons $\eta$',fontsize=15)


plt.xscale('log')
plt.yscale('log')

plt.savefig('BB.png',dpi=200,bbox_inches='tight')
```
