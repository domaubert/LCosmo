```python
th=np.arange(0,2*pcst.pi,0.01)
GM=1
E=-2
A=GM/2/np.abs(E)
B=np.sqrt(A**3/GM)
r=A*(1.-np.cos(th))
t=B*(th-np.sin(th))
v=A*np.sin(th)/(B*(1.-np.cos(th)))

Et=0.5*v**2-GM/r
tc=2*np.pi*B
ta=np.pi*B


d=4.5*(th-np.sin(th))**2/(1-np.cos(th))**3-1
dl=3./20.*(6*t/B)**(2./3.)
dc=np.copy(d)
w=np.where(d>=178)[0]
dc[w]=178
```

```python
plt.figure()
plt.plot(th,r,label=r'$r(\theta)$')
plt.plot(th,t,label=r'$t(\theta)$')
plt.xlabel(r'$\theta$',fontsize=15)
plt.legend(fontsize=15)
plt.savefig('rtparam.png')
```
