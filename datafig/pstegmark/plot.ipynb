```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10.0, 10.0) 
```

```python
th=np.loadtxt('th.txt',delimiter=";")
lya=np.loadtxt('lya.txt',delimiter=';')
lyam=lya[:11,]
lyap=lya[11:,]
sd=np.loadtxt('sdss.txt',delimiter=";")
sdm=sd[:16,]
sdp=sd[16:,]
wl=np.loadtxt('wl.txt',delimiter=";")
wlm=wl[:6,]
wlp=wl[6:,]
cmb=np.loadtxt('cmb.txt',delimiter=";")
cmm=cmb[:17,]
cmp=cmb[17:,]
```

```python
plt.figure()
plt.plot(th[:,0],th[:,1],label='P(k) théorique')
plt.fill_betweenx(lyam[:,1],lyam[:,0],lyap[:,0],color='0.1',alpha=0.1,hatch='x',label='Forêt LyA')
plt.fill_between(sdp[:,0],sdm[:,1],sdp[:,1],color='0.3',alpha=0.5,hatch='o',label='Galaxies SDSS')
plt.fill_between(wlp[:,0],wlm[:,1],wlp[:,1],color='0.6',alpha=0.5,label='Weak Lensing')
plt.fill_between(cmp[:,0],cmm[:,1],cmp[:,1],color='0.2',alpha=0.5,label='CMB',hatch='.')

plt.xscale('log')
plt.yscale('log')
plt.legend(fontsize=15)
plt.xlabel('fréquence spatiale [h/Mpc]',fontsize=15)
plt.ylabel(r'spectre de puissance actuel P(k) [Mpc/h]$^3$',fontsize=15)
plt.savefig('pstegmark.png',dpi=200,bbox_inches='tight')
```
